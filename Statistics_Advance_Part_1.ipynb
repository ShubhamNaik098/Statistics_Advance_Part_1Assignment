1.Independence:

    Sample observations must be independent of each other. This is usually ensured by random sampling or when the sample size is less than 10% of the population if sampling without replacement.

2.Sample Size:

    For populations that are not heavily skewed, a sample size n≥30 is generally sufficient.

    For highly skewed or non-normal populations, larger sample sizes may be needed.

3.Random Sampling:

    Samples should be drawn randomly from the population to avoid bias.

    The population from which samples are drawn should have a finite variance (finite standard deviation).

**Q9. Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero. Cannot compute Z-scores.")

    z_scores = [(x - mean) / std for x in data]
    plt.figure(figsize=(8, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue', label='Z-scores')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axhline(1, color='green', linestyle='--', linewidth=0.8, label='±1 SD')
    plt.axhline(-1, color='green', linestyle='--', linewidth=0.8)
    plt.axhline(2, color='orange', linestyle='--', linewidth=0.8, label='±2 SD')
    plt.axhline(-2, color='orange', linestyle='--', linewidth=0.8)
    plt.title('Z-Scores of Dataset')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-Score')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

data = np.random.normal(50, 10, 100)
z_scores = calculate_and_plot_z_scores(data)
```

**Q15.Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    reject_h0 = p_value < alpha
    return z_stat, p_value, reject_h0

np.random.seed(42)
population_std = 10
sample = np.random.normal(loc=52, scale=population_std, size=50)
population_mean = 50
alpha = 0.05

z_stat, p_val, reject = z_test(sample, population_mean, population_std, alpha)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print(f"Reject Null Hypothesis? {'Yes' if reject else 'No'}")
```

**Q16.Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=50)

ci_low, ci_high = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
print(f"This means we are 95% confident that the true population mean lies between {ci_low:.2f} and {ci_high:.2f}.")
```

**Q18.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mu=0, sigma=1):

    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    pdf_values = norm.pdf(x, loc=mu, scale=sigma)
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf_values, label=f'Normal PDF\nμ={mu}, σ={sigma}', color='blue')
    plt.fill_between(x, pdf_values, alpha=0.2, color='blue')
    plt.title('Probability Density Function of Normal Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

plot_normal_pdf(mu=0, sigma=1)
```

**Q21.Write a Python program to compare the standard deviations of two datasets and visualize the difference.**

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_stddev(data1, data2):
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    mean1 = np.mean(data1)
    mean2 = np.mean(data2)

    print(f"Dataset 1: Mean = {mean1:.2f}, Std Dev = {std1:.2f}")
    print(f"Dataset 2: Mean = {mean2:.2f}, Std Dev = {std2:.2f}")

    plt.figure(figsize=(10, 6))

    plt.hist(data1, bins=20, alpha=0.5, label='Dataset 1', color='blue', density=True)
    plt.axvline(mean1, color='blue', linestyle='-', linewidth=2, label='Mean Dataset 1')
    plt.axvspan(mean1 - std1, mean1 + std1, color='blue', alpha=0.2, label='±1 Std Dev Dataset 1')

    plt.hist(data2, bins=20, alpha=0.5, label='Dataset 2', color='orange', density=True)
    plt.axvline(mean2, color='orange', linestyle='-', linewidth=2, label='Mean Dataset 2')
    plt.axvspan(mean2 - std2, mean2 + std2, color='orange', alpha=0.2, label='±1 Std Dev Dataset 2')

    plt.title('Comparison of Standard Deviations for Two Datasets')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

np.random.seed(123)
data1 = np.random.normal(loc=50, scale=5, size=200)
data2 = np.random.normal(loc=55, scale=10, size=200)

compare_stddev(data1, data2)
```

**Q24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.**

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, Fisher’s definition (subtracts 3)
    return skewness, kurt

np.random.seed(42)
sample_data = np.random.normal(loc=0, scale=1, size=1000)

skewness, kurt = calculate_skewness_kurtosis(sample_data)
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis (Excess): {kurt:.4f}")
```
